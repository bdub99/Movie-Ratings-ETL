```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import the IMDB csv file
csv_file = "Resources/imdb_top_1000.csv"
imdb_df = pd.read_csv(csv_file)
imdb_df.head()
```

```python
#selecting only the columns that we want to focus on
imdb_df = imdb_df[["Series_Title", "Released_Year", "IMDB_Rating", "Meta_score", "Director", "Gross"]]
imdb_df
```

```python
#renaming column names to be lower cased to match the SQL file.
imdb_df.rename(columns={'Series_Title': 'title', 'Released_Year': 'year', 'Director': 'director', 'IMDB_Rating': 'imdb_rating', 'Meta_score': 'meta_score', 'Gross': 'gross'}, inplace=True)
imdb_df
```

```python
#import the Rotten Tomatoes csv file
csv_file = "Resources/rotten_tomatoes_top_movies.csv"
rotten_tomatoes_df = pd.read_csv(csv_file)
rotten_tomatoes_df.head()
```

```python
#selecting only the columns that we want to focus on.
rotten_tomatoes_df = rotten_tomatoes_df[["title", "year", "critic_score", "people_score", "genre", "director"]]
rotten_tomatoes_df
```

```python
#alphabetizing IMDB on title
imdb_df_sorted = imdb_df.sort_values('title')
imdb_df_sorted
```

```python
#dropping IMDB duplicates
imdb_df_sorted_dropped = imdb_df_sorted.dropna(subset = ['gross'], inplace = False)
imdb_df_sorted_dropped
```

```python
#alphabetizing Rotten Tomatoes on title
rotten_tomatoes_df_sorted = rotten_tomatoes_df.sort_values('title')
rotten_tomatoes_df_sorted
```

```python
#dropping Rotten Tomatoes duplicates
rotten_tomatoes_df_sorted_dropped = rotten_tomatoes_df_sorted.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
rotten_tomatoes_df_sorted_dropped
```

```python
connection_string = "postgres:postgres@localhost:5432/movies_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
imdb_df_sorted_dropped.to_sql(name='imdb', con=engine, if_exists='replace', index=True)
```

```python
rotten_tomatoes_df_sorted_dropped.to_sql(name='rotten_tomatoes', con=engine, if_exists='replace', index=True)
```
